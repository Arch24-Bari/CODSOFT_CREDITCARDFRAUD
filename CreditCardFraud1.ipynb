-----Bivariate Analysis------

```python
#population bins using qcut

s_train_df_resampled['pop_bin'] = pd.qcut(s_train_df_resampled['city_pop'], q=10, labels=False, duplicates='drop')

# Calculate the mean fraud rate for each bin
fraud_rate_by_pop_bin = s_train_df_resampled.groupby('pop_bin')['is_fraud'].mean().reset_index()

# Visualize the fraud rate by population bin
plt.figure(figsize=(10, 6))
sns.barplot(data=fraud_rate_by_pop_bin, x='pop_bin', y='is_fraud')
plt.title('Fraud Rate by City Population Bin')
plt.xlabel('City Population Bin')
plt.ylabel('Fraud Rate')
plt.xticks(rotation=45)
plt.show()

#Analysis = no clear linear relationship between city population and fraud rate.
```
